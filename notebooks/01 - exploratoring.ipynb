```python
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
imagem = Image.open('Images/image01.jpg')
imagem.thumbnail((256,256), Image.Resampling.LANCZOS)
imagem
```

```python
N, M = imagem.size
```

```python
x = np.asarray(imagem).reshape((N*M),3)
```

```python
x.shape
```

```python
model = KMeans(n_clusters=4).fit(x)

cores = model.cluster_centers_.astype('uint8')[np.newaxis]

plt.imshow(cores)
plt.axis('off')
plt.show()
```
